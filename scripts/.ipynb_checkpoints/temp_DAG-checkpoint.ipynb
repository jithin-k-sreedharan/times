## Find theta

```python
def find_theta(rank_new, rank_org, G_no_nodes):
    # rank_new is a dictionary as {nodeID: bin_no}
    # lower the bin_no the oldest the node is in the network. 
    # Multiple nodes can have same bin_no

    rank_new_ard = defaultdict(list)
    for k,v in rank_new.items():
        rank_new_ard[v].append(k)

    # Create a dictionary as {Bin_no: [list of nodeIDs]}
    # This is only slightly different from rank_new_ard when some consecutive -
    # bin_no's are not present.
    # rank_bin = {itn:rank_new_ard[val] for itn, val in enumerate(sorted(rank_new_ard))}
    
    rank_bin = rank_new_ard
    
    beta_tilde = 0
    rank_bin_len = len(rank_bin)
    for i in range(rank_bin_len-1):
        len_rank_bin_i = len(rank_bin[i])
        for j in range(i+1,rank_bin_len):
            len_rank_bin_j = len(rank_bin[j])
            beta_ij = 0;
            for u in rank_bin[i]:
                tt = np.array([rank_org[ttt] for ttt in rank_bin[j]])
                beta_ij += sum(rank_org[u] < tt)
            beta_ij /= (len_rank_bin_i*len_rank_bin_j)
            beta_tilde += beta_ij

    eta = beta_tilde * 2/((rank_bin_len-1)*G_no_nodes)
    eta_1 =  beta_tilde *2/((rank_bin_len-1)*rank_bin_len)
    return (eta,eta_1)
```

```python
# the level of v is the length of the longest directed path from v
def find_levels_DAG_long_path_t(G, v, level):
    if v not in level:
        if v not in G.nodes() or not G.neighbors(v):
            level[v] = 0
        else:
            level[v] = max(find_levels_DAG_long_path_t(G, w, level) + 1 for w in G.neighbors(v))
    return level[v]

def find_levels_DAG_long_path(G):
    l = {}
    for v in G.nodes():
        find_levels_DAG_long_path_t(G, v, l)

    l_val_s = sorted(set(l.values()))
    l_val_new = dict(zip(l_val_s,range(len(l_val_s))))
    l_new = {k:l_val_new[v] for k,v in l.items()}
    
    rank_new_bin = defaultdict(list)
    for k,v in l_new.items():
        rank_new_bin[v].append(k)
    return (l_new,rank_new_bin)
```

```python
def find_levels_DAG_ISIT(G):
    # This algorithm finds the levels of the DAG
    G_n =  G.copy()
    i = 0
    rank_new = defaultdict(int)
    rank_new_bin = defaultdict(list)
    while(G_n.number_of_nodes() > 0):
        degree_list_ard = defaultdict(list)
        for k,v in G_n.in_degree().items():
            degree_list_ard[v].append(k)
            
        rank_bin_0 = degree_list_ard[0]
        rank_new_bin[i] = rank_bin_0
        
        rank_new.update(dict.fromkeys(rank_bin_0,i))
        G_n.remove_nodes_from(rank_bin_0)
        i += 1

    depth_dag = i
    rank_new = {k:((depth_dag-1)-rank_new[k]) for k in rank_new.keys()}
    rank_new_bin = {(depth_dag-1)-k:v for k,v in rank_new_bin.items()}

    return (rank_new,rank_new_bin)
```

```python
def find_seq_rank_unif(G):
    # This is Sequential technique (uniform)
    G_n =  G.copy()
    i = G_n.number_of_nodes()-1
    rank_new = defaultdict(int)

    while(i > 0):
     
        # degree_list_ard is a dictionary with {degree:[list of nodes with this degree]
        degree_list_ard = defaultdict(list)
        for k,v in G_n.degree().items():
            degree_list_ard[v].append(k)
            
        rank_bin_0 = degree_list_ard[min(degree_list_ard)]
#         print(rank_bin_0)
        node_sel =  np.random.choice(rank_bin_0)

        rank_new[node_sel] = i
        G_n.remove_node(node_sel)
        i -= 1
    rank_new[G_n.nodes()[0]] = i
    return rank_new
```

```python
#parameters
no_runs = 1000
time_n = 100
pr_alpha = 1
p_lb = 5
p_ub = 5
if (p_lb != p_ub):
    vec_p = np.array([np.arange(p_lb,p_ub),(1/(p_ub-p_lb)*np.ones((p_ub-p_lb)))])#, dtype = object)
else:
    vec_p = np.array([[p_lb],[1]])
pr_beta = 0.25
if (p_lb != p_ub):
    vec_q = np.array([np.arange(p_lb,p_ub),(1/(p_ub-p_lb)*np.ones((p_ub-p_lb)))])#, dtype = object)
else:
    vec_q = np.array([[p_lb],[1]])
pr_delta = 0.5
pr_gamma = 0.5

G = nx_jks.preferential_attachement_genzd(time_n, pr_alpha, vec_p, pr_beta,\
                                               vec_q, pr_delta, pr_gamma, directed = True)
rank_orig = dict(zip(sorted(G.nodes()),range(G.number_of_nodes())))

rank_seq_pref = find_seq_rank_unif(G)
(theta_seq_pref_t, theta_seq_pref_t_1) = find_theta(rank_seq_pref,rank_orig,G.number_of_nodes())
```

## find_parallel_rank

```python
def find_parallel_rank(G):
    # This is Parallel technique
    G_n =  G.copy()
    i = 0
    rank_new = defaultdict(int)
    rank_new_bin = defaultdict(list)
    while(G_n.number_of_nodes() > 0):
        degree_list_ard = defaultdict(list)
        for k,v in G_n.degree().items():
            degree_list_ard[v].append(k)
            
        rank_bin_0 = degree_list_ard[min(degree_list_ard)]
#         print(rank_bin_0)
        rank_new_bin[i] = rank_bin_0

        rank_new.update(dict.fromkeys(rank_bin_0,i))
        G_n.remove_nodes_from(rank_bin_0)
        i += 1

    depth_dag = i
    rank_new = {k:((depth_dag-1)-rank_new[k]) for k in rank_new.keys()}
    rank_new_bin = {(depth_dag-1)-k:v for k,v in rank_new_bin.items()}
    
    return (rank_new,rank_new_bin)
```

## find_RW_based_ranking

```python
def find_SRW_tour_length(G,v):
    neighbors = nx.neighbors(G,v)
    sample = np.random.choice(neighbors)
    xi = 1
    while sample != v:
        neighbors = nx.neighbors(G,sample)
        sample = np.random.choice(neighbors)
        xi += 1
#         print("v={0},sample={1}".format(v,sample))
    return xi

def find_NB_RW_tour_length(G,v):
    neighbors = nx.neighbors(G,v)
    try:
        neighbors.remove(v)
    except ValueError:
        pass
    sample_p = v
#     sample = max(G.degree(neighbors), key = G.degree(neighbors).get)
    sample = np.random.choice(neighbors)
    xi = 1
    while sample != v:
        neighbors = nx.neighbors(G,sample)
        try:
            neighbors.remove(sample)
        except ValueError:
            pass
        try:
            neighbors.remove(sample_p)
        except ValueError:
            pass
        sample_p = sample        
#         sample = max(G.degree(neighbors), key = G.degree(neighbors).get)        
        sample = np.random.choice(neighbors)
        xi += 1
        print("v={0},sample={1}".format(v,sample))    
    return xi

def find_RW_based_ranking(G):
    rank_RW_t = {}
    for v in G.nodes():
#         rank_RW_t[v] = find_SRW_tour_length(G,v)
        rank_RW_t[v] = find_NB_RW_tour_length(G,v)
    rank_RW_t_val_s = sorted(set(rank_RW_t.values()))
    rank_RW_t_val_new = dict(zip(rank_RW_t_val_s,range(len(rank_RW_t_val_s))))
    rank_RW = {k:rank_RW_t_val_new[v] for k,v in rank_RW_t.items()}
    
    rank_RW_bin = defaultdict(list)
    for k,v in rank_RW.items():
        rank_RW_bin[v].append(k)
    return (rank_RW,rank_RW_bin)
```

# Finding Adam

## Finding Adam [Bubeck et al]

```python
#parameters
time_n = 100
pr_alpha = 1
p_lb = 1
p_ub = 1
if (p_lb != p_ub):
    vec_p = np.array([np.arange(p_lb,p_ub),(1/(p_ub-p_lb)*np.ones((p_ub-p_lb)))])#, dtype = object)
else:
    vec_p = np.array([[p_lb],[1]])
pr_beta = 1
if (p_lb != p_ub):
    vec_q = np.array([np.arange(p_lb,p_ub),(1/(p_ub-p_lb)*np.ones((p_ub-p_lb)))])#, dtype = object)
else:
    vec_q = np.array([[p_lb],[1]])
pr_delta = 0.6
pr_gamma = 0.6
```

```python
def dfs_subtree_size_max(G,root):

    tree = nx.dfs_tree(G, root)
    tree.remove_node(root)
 
    res = 1
    for node in tree:
        tt = len(nx.dfs_tree(tree,node))
        if tt > res:
            res = tt

    return res

adam_paper_rank_0 ={v:dfs_subtree_size_max(G,v) for v in G.nodes()}
adam_paper_rank_0_s_k = sorted(adam_paper_rank_0, key = adam_paper_rank_0.get)
```

```python
adam_paper_rank_0_s_k
```

```python
def dfs_subtree_size_prod(G,root):
    tree = nx.dfs_tree(G, root)
    tree.remove_node(root)

    prod = 1
    for node in tree:
        prod *= len(nx.dfs_tree(tree,node))
        
    return prod

adam_paper_rank ={v:dfs_subtree_size_prod(G,v) for v in G.nodes()}
adam_paper_rank_s_k = sorted(adam_paper_rank, key = adam_paper_rank.get)
```

```python
# %%cython
# cimport numpy as np
# cimport networkx as nx
# # cimport random
@jit
def preferential_attachement_genzd(time_n, pr_alpha, vec_p, pr_beta, vec_q, pr_delta, pr_gamma, directed = False):
    """Returns a random graph sample of generalized preferential attachement model

    Parameters
    ----------
    time_n
    pr_alpha: probability for adding a new node. (1-pr_alpha): prob for adding edges between existing nodes
    vec_p: Distrbution for m when a node is added; first row is values, second row is probability vector
    pr_beta: probability that edges to new node selected preferentially. (1-pr_beta): uniformly adding edges to new node
    vec_q: Distrbution for m when edges between existing nodes added; first row is values, second row is probability vector
    pr_delta: Adding edges between exisiting nodes, Source node selected preferentially
    pr_gamma: Adding edges between exisiting nodes, Terminal node selected preferentially
    seed = None
    directed = False

    Return
    -------
    G: Networkx Graph

    Notes
    -----
    While adding "m" edges to the network (new node or between exisiting nodes) at time n, the degrees of the network at (n-1)
    are used, not updated degrees.

    References
    ----------
    [Cooper and Freize]
    """

    # Make a list of exceptions
    # if m < 1 or  m >=n:
    #     raise nx.NetworkXError(\
    #           "Barabasi-Albert network must have m>=1 and m<n, m=%d,n=%d"%(m,n))

    #Random seed generation
#     random.seed()
    np.random.seed()

    # Create an empty MultiGraph which allows multiple self loops and multi edges
    if directed == True:
        G = nx.MultiDiGraph()
    else:
        G = nx.MultiGraph()

    # Add two intial nodes
    G.add_node(0)
    rv_m_new = np.random.choice(vec_p[0,:].astype(int),p = vec_p[1,:])
    G.add_edges_from([(0,0)]*rv_m_new)
    target_list = [0]*2*rv_m_new
    G.add_node(1)
    rv_m_new = np.random.choice(vec_p[0,:].astype(int),p = vec_p[1,:])
    G.add_edges_from([(1,0)]*rv_m_new)
    target_list.extend([0]*rv_m_new + [1]*rv_m_new)

    source = 2
    time_i = 2
    while time_i < time_n:
        # rv_alpha: 1, new node; 0, existing node
        rv_alpha = np.random.binomial(1,pr_alpha)
        nodes_prev_time = G.nodes()
        if rv_alpha:
            # new node
            rv_m_new = np.random.choice(vec_p[0,:].astype(int),p = vec_p[1,:])
            target_list_temp = []
            for e in range(rv_m_new):
                rv_beta =  np.random.binomial(1,pr_beta)
                if rv_beta:
                    # Edges to new node selected preferentially
                    target = np.random.choice(target_list)
                    G.add_edge(source,target)
                    target_list_temp.extend([source,target])
                else:
                    # Terminal node of edge to new node is selected at uniformly at random
                    target = np.random.choice(nodes_prev_time)
                    G.add_edge(source,target)
                    target_list_temp.extend([source,target])
            source += 1
            target_list.extend(target_list_temp)
        else:
            # old node
            rv_m_old = np.random.choice(vec_q[0,:].astype(int),p = vec_q[1,:])
            target_list_temp = []
            for e in range(rv_m_old):
                rv_delta =  np.random.binomial(1,pr_delta)
                rv_gamma =  np.random.binomial(1,pr_gamma)
                if rv_delta:
                    # Source node selected preferentially
                    source_exist = np.random.choice(target_list)
                else:
                    # Source node selected uniformly
                    source_exist = np.random.choice(nodes_prev_time)

                if rv_gamma:
                    # Terminal node selected preferentially
                    target = np.random.choice(target_list)
                else:
                    # Terminal node selected uniformly
                    target = np.random.choice(nodes_prev_time)

                G.add_edge(source_exist,target)
                target_list_temp.extend([source_exist,target])
            target_list.extend(target_list_temp)
        time_i += 1
    return G
```
